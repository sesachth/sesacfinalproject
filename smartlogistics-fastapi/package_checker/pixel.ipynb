```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# 이미지 데이터 경로
image_dir = r"C:/labs_python/src_cnn_data/output_image/box_damaged_1412"

# 이미지 데이터를 저장할 리스트
all_pixels = []
resolutions = {}

# 디렉터리 내 모든 이미지 파일 읽기
for filename in os.listdir(image_dir):
    if filename.lower().endswith(('.jpg', '.png', '.jpeg')):
        image_path = os.path.join(image_dir, filename)
        try:
            with Image.open(image_path) as img:
                img_array = np.array(img)
                all_pixels.append(img_array.flatten())
                # 해상도 저장
                resolution = img.size
                if resolution in resolutions:
                    resolutions[resolution] += 1
                else:
                    resolutions[resolution] = 1
        except Exception as e:
            print(f"Error processing {filename}: {e}")

# 해상도 범위 세분화 그룹화
resolution_bins = {
    '0-100x0-100': 0,
    '101-200x101-200': 0,
    '201-300x201-300': 0,
    '301-400x301-400': 0,
    '401-500x401-500': 0,
    '501-600x501-600': 0,
    '601+': 0
}

for res, count in resolutions.items():
    if res[0] <= 100 and res[1] <= 100:
        resolution_bins['0-100x0-100'] += count
    elif res[0] <= 200 and res[1] <= 200:
        resolution_bins['101-200x101-200'] += count
    elif res[0] <= 300 and res[1] <= 300:
        resolution_bins['201-300x201-300'] += count
    elif res[0] <= 400 and res[1] <= 400:
        resolution_bins['301-400x301-400'] += count
    elif res[0] <= 500 and res[1] <= 500:
        resolution_bins['401-500x401-500'] += count
    elif res[0] <= 600 and res[1] <= 600:
        resolution_bins['501-600x501-600'] += count
    else:
        resolution_bins['601+'] += count

# 해상도 그룹화 시각화
plt.figure(figsize=(10, 6))
plt.bar(resolution_bins.keys(), resolution_bins.values(), color='orange')
plt.title('Binned Image Resolution Distribution')
plt.xlabel('Resolution Range')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 이미지 파일 확인 및 픽셀 값 시각화
if all_pixels:  
    all_pixels = np.concatenate(all_pixels)
    all_pixels_normalized = all_pixels / 255.0

    plt.figure(figsize=(8, 6))
    plt.hist(all_pixels_normalized, bins=50, color='blue', alpha=0.7)
    plt.title('Pixel Value Distribution (Normalized)')
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
else:
    print("No valid image files found in the directory.")
```
