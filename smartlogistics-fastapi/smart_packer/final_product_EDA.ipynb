```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rc('font', family='gulim')
```

```python
# 엑셀 파일 경로
file_path = '물품_정보_세부_카테고리_분류.xlsx'

# 엑셀 파일을 DataFrame으로 읽기
df = pd.read_excel(file_path)

# DataFrame 확인
print(df)
```

```python
df.info()
```

```python
df.describe()
```

```python
# 카테고리 별 제품 개수 계산
category_counts = df['세부_카테고리'].value_counts()

# 막대 그래프 그리기
plt.bar(category_counts.index, category_counts.values)

plt.xlabel('세부 카테고리')
plt.ylabel('제품 개수')
plt.title('세부 카테고리 별 제품 개수 분포')
plt.xticks(rotation=45)
plt.show()
```

```python
# 히스토그램 그리기
plt.hist(df['가로(mm)'], bins=range(0, int(df['가로(mm)'].max()), 100), edgecolor='black')  # bins는 구간 수
plt.xlabel('가로(mm)')
plt.ylabel('Frequency')
plt.title('Histogram of 가로(mm)')
plt.grid()
plt.show()

# 상자 플롯 그리기
plt.figure(figsize=(10, 2))
plt.boxplot(df['가로(mm)'], vert=False)  # vert=False로 설정하여 수평 방향으로 그리기

# 그래프 꾸미기
plt.title('가로(mm) Box Plot')
plt.xlabel('가로(mm)')
plt.ylabel('Dataset')
```

```python
# 히스토그램 그리기
plt.hist(df['세로(mm)'], bins=range(0, int(df['세로(mm)'].max()), 100), edgecolor='black')  # bins는 구간 수
plt.xlabel('세로(mm)')
plt.ylabel('Frequency')
plt.title('Histogram of 세로(mm)')
plt.grid()
plt.show()

# 상자 플롯 그리기
plt.figure(figsize=(10, 2))
plt.boxplot(df['세로(mm)'], vert=False)  # vert=False로 설정하여 수평 방향으로 그리기

# 그래프 꾸미기
plt.title('세로(mm) Box Plot')
plt.xlabel('세로(mm)')
plt.ylabel('Dataset')
```

```python
plt.figure(figsize=(20, 10))

# 박스 플롯
sns.boxplot(x="세부_카테고리", y="중량(g)", data=df)

# 바이올린 플롯
sns.violinplot(x="세부_카테고리", y="중량(g)", data=df)

plt.yticks(np.arange(0, 20000, 1000))
plt.title("세부 카테고리 별 중량 분포도")
plt.show()
```

```python
# 취급주의 값이 1인 데이터만 필터링
fragile_data = df[df['취급주의'] == 1]

# 카운터 플롯
ax = sns.countplot(x='세부_카테고리', data=fragile_data)
plt.title('세부 카테고리 별 취급주의 물품 개수')
plt.xlabel('세부_카테고리')
plt.ylabel('취급주의 물품 개수')

# 각 막대 위에 카운트 값 표시
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2., height + 0.1,
            f'{height:.0f}',
            ha="center", va="bottom")
    
plt.show()
```

```python
# 산점도 행렬
for category in df['세부_카테고리'].unique():
    category_data = df[df['세부_카테고리'] == category][['가로(mm)', '세로(mm)', '높이(mm)']]
    
    sns.pairplot(category_data)
    plt.suptitle(f'Scatter Plot Matrix for {category}')
    plt.show()
```

```python
# 히트맵
corr_matrix = df[['가로(mm)', '세로(mm)', '높이(mm)', '중량(g)']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('크기와 중량에 대한 상관관계 히트맵')
plt.show()
```

```python
# 데이터프레임에 '크기_합계' 열 추가
df['크기_합계'] = df['가로(mm)'] + df['세로(mm)'] + df['높이(mm)']

# 500mm 미만인 물품만 필터링
df_small = df[df['크기_합계'] < 500]

# 카테고리별로 개수 계산
category_counts = df_small['세부_카테고리'].value_counts().reset_index()
category_counts.columns = ['세부_카테고리', '개수']

# 카운트 플롯 그리기
plt.figure(figsize=(12, 6))
sns.barplot(x='세부_카테고리', y='개수', data=category_counts)
plt.title('가로/세로/높이 총 합이 500mm 미만인 물품 개수')
plt.xlabel('세부_카테고리')
plt.ylabel('개수')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# 각 막대 위에 개수 표시
for i, v in enumerate(category_counts['개수']):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```
